```python
def check_permission(user_attr, resource_attr):
    # print(user_attr)
    rules = [
        {"id": 1, "condition": lambda: "manager" in user_attr["admin_roles"] and resource_attr["type"] == "budget" and resource_attr["activity"] in {"read", "approve"} and resource_attr["rsc_department"] == user_attr["usr_department"]},
        {"id": 2, "condition": lambda: resource_attr["type"] in {"schedule", "budget"} and resource_attr["activity"] in {"read", "write"} and resource_attr["rsc_prj"] == user_attr["prj_led"]},
        {"id": 3, "condition": lambda: resource_attr["type"] == "schedule" and resource_attr["activity"] == "read" and resource_attr["rsc_prj"] in user_attr["usr_prj"]},
        {"id": 4, "condition": lambda: resource_attr["type"] == "task" and resource_attr["activity"] == "setStatus" and resource_attr["task"] == user_attr["tasks"]},
        {"id": 5, "condition": lambda: resource_attr["type"] == "task" and not resource_attr["proprietary"] and resource_attr["activity"] in {"request", "read"} and resource_attr["rsc_prj"] in user_attr["usr_prj"] and user_attr["usr_expertise"] is not None and resource_attr["rsc_expertise"] is not None and user_attr["usr_expertise"] < resource_attr["rsc_expertise"]},
        {"id": 6, "condition": lambda: user_attr["isEmployee"] and resource_attr["type"] == "task" and resource_attr["activity"] in {"request", "read"} and resource_attr["rsc_prj"] in user_attr["usr_prj"] and user_attr["usr_expertise"] is not None and resource_attr["rsc_expertise"] is not None and user_attr["usr_expertise"] < resource_attr["rsc_expertise"]},
        {"id": 7, "condition": lambda: "auditor" in user_attr["admin_roles"] and resource_attr["type"] == "budget" and resource_attr["activity"] == "read" and resource_attr["rsc_prj"] in user_attr["usr_prj"]},
        {"id": 8, "condition": lambda: "accountant" in user_attr["admin_roles"] and resource_attr["type"] == "budget" and resource_attr["activity"] in {"read", "write"} and resource_attr["rsc_prj"] in user_attr["usr_prj"]},
        {"id": 9, "condition": lambda: "accountant" in user_attr["admin_roles"] and resource_attr["type"] == "task" and resource_attr["activity"] == "setCost" and resource_attr["rsc_prj"] in user_attr["usr_prj"]},
        {"id": 10, "condition": lambda: "planner" in user_attr["admin_roles"] and resource_attr["type"] == "schedule" and resource_attr["activity"] == "write" and resource_attr["rsc_prj"] in user_attr["usr_prj"]},
        {"id": 11, "condition": lambda: "planner" in user_attr["admin_roles"] and resource_attr["type"] == "task" and resource_attr["activity"] == "setSchedule" and resource_attr["rsc_prj"] in user_attr["usr_prj"]},
    ]

    for rule in rules:
        if rule["condition"]():
            return True
    
    return False

def random_entry(entry_type, admin_role, projects_pool, tasks_pool, departments_pool):
    user_id = str(uuid.uuid4())

    curr_usr_prj = random.choice(projects_pool)

    user_attr = {
        'user_id': user_id,
        'usr_prj': curr_usr_prj,
        'prj_led': curr_usr_prj if random.random() < 0.5 else None,
        'admin_roles': admin_role,
        'usr_expertise': random.choice(expertise_areas) if random.random() < 0.5 else None,
        'tasks': random.choice(tasks_pool),
        'usr_department': random.choice(departments_pool) if random.random() < 0.5 else None,
        'isEmployee': random.choice([True, False])
    }

    resource_attr = {
        'type': random.choice(resource_types),
        'rsc_prj': random.choice(projects_pool),
        'rsc_department': random.choice(departments_pool),
        'rsc_expertise': random.choice(expertise_areas) if random.random() < 0.5 else None,
        'proprietary': random.choice([True, False]),
        'task': random.choice(tasks_pool),
        'activity': random.choice(resource_actions)
    }

    permitted = check_permission(user_attr, resource_attr)

    # Adjust the permitted flag based on the entry_type
    if entry_type == EntryType.ONLY_PERMITTED:
        if not permitted:
            resource_attr['proprietary'] = False
            permitted = True
    elif entry_type == EntryType.ONLY_DENIED:
        if permitted:
            resource_attr['proprietary'] = True
            permitted = False

    return {**user_attr, **resource_attr, 'action': permitted}

def generate_and_save_entries(n, entry_type, file_path, admin_role, projects_pool, tasks_pool, departments_pool):
    with open(file_path, 'a', newline='') as output_file:
        fieldnames = ['user_id', 'usr_prj', 'prj_led', 'admin_roles', 'usr_expertise', 'tasks', 'usr_department', 'isEmployee',
                      'type', 'rsc_prj', 'rsc_department', 'rsc_expertise', 'proprietary', 'task', 'activity', 'action']
        dict_writer = csv.DictWriter(output_file, fieldnames=fieldnames)
        
        # Write the header only once if the file is empty
        if output_file.tell() == 0:
            dict_writer.writeheader()

        for _ in range(n):
            entry = random_entry(entry_type, admin_role, projects_pool, tasks_pool, departments_pool)
            dict_writer.writerow(entry)
```

```python
import os
import glob

def delete_files_matching_pattern(directory, pattern):
    """
    Delete all files in the specified directory matching the given pattern.
    
    :param directory: The directory where the files are located.
    :param pattern: The pattern to match the files to be deleted.
    """
    # Construct the full path with the pattern
    full_path_pattern = os.path.join(directory, pattern)

    # Find all files matching the pattern
    files_to_delete = glob.glob(full_path_pattern)

    # Delete each file
    for file_path in files_to_delete:
        try:
            os.remove(file_path)
            print(f"Deleted: {file_path}")
        except Exception as e:
            print(f"Error deleting {file_path}: {e}")
```

```python
def extract_digits(value):
    if isinstance(value, int):
        return value
        
    match = re.search(r'\d+', value)
    if match:
        return int(match.group())
    return None

def extract_digits_flt(value):
    if isinstance(value, int):
        return value
        
    return int(value)

def extract_bools(value):
    if value:
        return int(1)
    return int(0)

def extract_and_format(value):
    if isinstance(value, int):
        return value
    # Extract digits from the string
    digits = ''.join(re.findall(r'\d+', value))
    
    # Check the suffix and append appropriate zeros
    if value.endswith('propa'):
        result = digits + '00'
    elif value.endswith('a'):
        result = digits + '0'
    else:
        result = digits
    
    # Convert the result to an integer
    return int(result)
    
    return None

def extract_and_format_admin_roles(value):
    if isinstance(value, int):
        return value
    
    if value == 'manager':
        return 10
    if value == 'accountant':
        return 11
    if value == 'auditor':
        return 12
    if value == 'planner':
        return 13

def extract_and_format_types(value):
    if isinstance(value, int):
        return value
    
    if value == 'budget':
        return 20
    if value == 'schedule':
        return 21
    if value == 'task':
        return 22

def extract_and_format_actions(value):
    if isinstance(value, int):
        return value
    
    if value == 'read':
        return 30
    if value == 'approve':
        return 31
    if value == 'write':
        return 32
    if value == 'setStatus':
        return 33
    if value == 'request':
        return 34
    if value == 'setCost':
        return 35
    if value == 'setSchedule':
        return 36
```
